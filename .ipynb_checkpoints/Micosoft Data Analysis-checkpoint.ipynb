```python
import pandas as pd

mp_df=pd.read_csv('Microsoft_data.csv')

mp_df
```

```python
mp_df.shape
```

```python
mp_df['OsBuild']=mp_df['OsBuild'].astype('object')

mp_df['HasDetections']=mp_df['HasDetections'].astype('object')
```

# Chi Squared Test

```python
from scipy.stats import chi2_contingency   #website:scipy.stats

pd.crosstab(mp_df['OsBuild'],mp_df['HasDetections'])  #observed table
```

```python
from scipy.stats import chi2_contingency    

contingency_table=pd.crosstab(mp_df['OsBuild'],mp_df['HasDetections'])  #observed table

p_value=chi2_contingency(contingency_table) 

p_value
```

```python
from scipy.stats import chi2_contingency

contingency_table=pd.crosstab(mp_df['OsBuild'],mp_df['HasDetections'])  #observed table

p_value=chi2_contingency(contingency_table)[1]  # p-value=o.oooo.....143411<0.05(S.L),
                                                #Reject H0(correlation exist)

p_value
```

```python
mp_df.columns

mp_df['HasDetections']=mp_df['HasDetections'].astype('object')

core=[]

for col in mp_df.columns:
    mp_df[col]=mp_df[col].astype('object')
    contingency_table=pd.crosstab(mp_df[col],mp_df['HasDetections'])
    p_value=chi2_contingency(contingency_table)[1]
    if p_value<0.05:
        core.append(col)
print(core)


        
    
```
